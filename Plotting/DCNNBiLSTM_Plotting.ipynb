```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/Maruata/Desktop/model_dcnnbilstm.csv")
```

```python
sn.set(font_scale=2)

history_dict = data
dnn_cicids2018_accuracy = history_dict['dnn_accuracy']
dnn_cicids2018_val_accuracy = history_dict['dnn_val_accuracy']
dnn_cicids2018_loss = history_dict['dnn_loss']
dnn_cicids2018_val_loss = history_dict['dnn_val_loss']

cnn_cicids2018_accuracy = history_dict['cnn_accuracy']
cnn_cicids2018_val_accuracy = history_dict['cnn_val_accuracy']
cnn_cicids2018_loss = history_dict['cnn_loss']
cnn_cicids2018_val_loss = history_dict['cnn_val_loss']

ae_cicids2018_accuracy = history_dict['ae_accuracy']
ae_cicids2018_val_accuracy = history_dict['ae_val_accuracy']
ae_cicids2018_loss = history_dict['ae_loss']
ae_cicids2018_val_loss = history_dict['ae_val_loss']

lstm_cicids2018_accuracy = history_dict['lstm_accuracy']
lstm_cicids2018_val_accuracy = history_dict['lstm_val_accuracy']
lstm_cicids2018_loss = history_dict['lstm_loss']
lstm_cicids2018_val_loss = history_dict['lstm_val_loss']

dcnnbilstm_cicids2018_accuracy = history_dict['dcnnbilstm_accuracy']
dcnnbilstm_cicids2018_val_accuracy = history_dict['dcnnbilstm_val_accuracy']
dcnnbilstm_cicids2018_loss = history_dict['dcnnbilstm_loss']
dcnnbilstm_cicids2018_val_loss = history_dict['dcnnbilstm_val_loss']

start_by_epoch = 1
epochs = range(start_by_epoch, len(dnn_cicids2018_accuracy) + 1)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, dnn_cicids2018_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dnn_cicids2018_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, cnn_cicids2018_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, cnn_cicids2018_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, ae_cicids2018_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, ae_cicids2018_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, lstm_cicids2018_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, lstm_cicids2018_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnnbilstm_cicids2018_accuracy[start_by_epoch-1:], 'x--', linewidth=3)
plt.plot(epochs, dcnnbilstm_cicids2018_val_accuracy[start_by_epoch-1:], 'x--', linewidth=3)


plt.title('Comparison of Training and Testing Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.legend(['Accuracy: DNN Trained with CICIDS2018', 'Validation Accuracy: DNN Trained with CICIDS2018', 
           'Accuracy: CNN Trained with CICIDS2018', 'Validation Accuracy: CNN Trained with CICIDS2018', 
           
           'Accuracy: AE Trained with CICIDS2018', 'Validation Accuracy: AE Trained with CICIDS2018', 
           'Accuracy: LSTM Trained with CICIDS2018', 'Validation Accuracy: LSTM Trained with CICIDS2018',
           
           'Accuracy: DCNNBiLSTM Trained with CICIDS2018', 'Validation Accuracy: DCNNBiLSTM Trained with CICIDS2018'
           ], loc='lower right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0.77,1.012)

plt.show()

fig.savefig('DCNNBiLSTM_Training_Validation_Accuracy.pdf') 
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, dnn_cicids2018_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dnn_cicids2018_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, cnn_cicids2018_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, cnn_cicids2018_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, ae_cicids2018_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, ae_cicids2018_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, lstm_cicids2018_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, lstm_cicids2018_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnnbilstm_cicids2018_loss[start_by_epoch-1:], 'x--', linewidth=3)
plt.plot(epochs, dcnnbilstm_cicids2018_val_loss[start_by_epoch-1:], 'x--', linewidth=3)


plt.title('Comparison of Training and Testing Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss Rate')

plt.legend(['Loss: DNN Trained with CICIDS2018', 'Validation Loss: DNN Trained with CICIDS2018', 
           'Loss: CNN Trained with CICIDS2018', 'Validation Loss: CNN Trained with CICIDS2018', 
           
           'Loss: AE Trained with CICIDS2018', 'Validation Loss: AE Trained with CICIDS2018', 
           'Loss: LSTM Trained with CICIDS2018', 'Validation Loss: LSTM Trained with CICIDS2018',
           
           'Loss: DCNNBiLSTM Trained with CICIDS2018', 'Validation Loss: DCNNBiLSTM Trained with CICIDS2018'
           ], loc='upper right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
#plt.ylim(0.77,1.012)

plt.show()

fig.savefig('DCNNBiLSTM_Training_Validation_Loss.pdf') 
```
